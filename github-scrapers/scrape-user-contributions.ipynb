```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
df=pd.read_csv('user-list.csv')
```

```python
df.head()
```

```python
namelist=[]
for name in df['github_link']:
    namelist.append(name)
```

```python
namelist
```

```python
activities=[]

for url in namelist:
    r=requests.get(url).text
    data=BeautifulSoup(r,"html.parser")
    information=data.find_all('rect')
    for j in range(217,len(information)): #start from July 1st
        activity={}
        activity['url'] = url
        activity['data-date'] = information[j]['data-date']
        activity['data-count'] = information[j]['data-count']
        activities.append(activity)
```

```python
#activities
```

```python
df2 = pd.DataFrame(activities)
```

```python
df2.head()
```

```python
#data-count is not number but string
df3 = df2.groupby('url')['data-count'].agg('sum').reset_index()
df3
```

```python
df2['data-count'] = pd.to_numeric(df2['data-count'])
df3 = df2.groupby('url')['data-count'].agg('sum').reset_index()
```

```python
df3.rename(columns={'url':'github_link'}, inplace=True)
```

```python
#merge two df into one, they must at have one sharing column
df4 = pd.merge(df,df3,how='left',on='github_link')
```

```python
df5 = df4.sort_values(by=['data-count'],ascending=False)
```

```python
def get_id(x):
    return x.split('https://github.com/')[-1]
```

```python
df5['github_id']=df['github_link'].apply(get_id)
```

```python
df5
```

```python
df5.to_csv('../data/activity_counts.csv', index=None)
```
