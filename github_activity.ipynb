```python
import pandas as pd
import requests
import time
df=pd.read_csv('activity_counts.csv')
```

```python
users = df['github_id'].tolist()
```

```python
under_100=users[6:] #filter out students' commits under 100
```

```python
under_100
```

```python
def get_under_100(user):
    url = 'https://api.github.com/users/{0}/events/public?page=1&per_page=100'.format(user)
    r = requests.get(url,headers={
    'User-Agent':'Mozilla/5.0',
    'Authorization':'token 959cfcb7ec90d04eab4968bf91862f618b347a94',
    'Content-Type':'application/json',
    'method':'GET',
    'Accept':'application/json'
    })
    content = r.json()
    commits = []
    for item in content:
        commit = {}
        commit['time'] = item['created_at']
        commit['repo'] = item['repo']['name']
        commit['users'] = item['actor']['login']
        commit['type'] = item['type']
        commits.append(commit)
    return commits
```

```python
for user in under_100:
    commits = get_under_100(user)
```

```python
all_commits=[]

for user in under_100:
    time.sleep(2)
    try:
        commits = get_under_100(user)
        all_commits.extend(commits)
    except:
        print(user)
```
